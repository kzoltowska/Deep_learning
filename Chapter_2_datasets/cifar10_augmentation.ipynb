```python
import numpy as np
from PIL import Image
```

```python
def augment(im, dim):
    img=Image.fromarray(im)
    if (np.random.random()<0.5):
        img=img.transpose(Image.FLIP_LEFT_RIGHT)
    if (np.random.random()<0.33333):
        z=(32-dim)/2
        r=10*np.random.random()-5
        img.rotate(r, resample=Image.BILINEAR)
        img=img.crop((z,z,32-z,32-z))
    else:
        x=int((32-dim-1)*np.random.random())
        y=int((32-dim-1)*np.random.random())
        img=img.crop((x,y,x+dim,y+dim))
    return np.array(img)

def main():
    x=np.load("/home/kasia/Deep_learning/Datasets/cifar10_train_images.npy")
    y=np.load("/home/kasia/Deep_learning/Datasets/cifar10_train_labels.npy")
    factor=10
    dim=28
    z=(32-dim)/2
    newx=np.zeros((x.shape[0]*factor, dim, dim, 3), dtype="uint8")
    newy=np.zeros(y.shape[0]*factor, dtype="uint8")
    k=0

    for i in range(x.shape[0]):
        im=Image.fromarray(x[i,:])
        im=im.crop((z,z,32-z, 32-z))
        newx[k,...]=np.array(im)
        newy[k]=y[i]
        k+=1
        for j in range(factor-1):
            newx[k,...]=augment(x[i,:], dim)
            newy[k]=y[i]
            k+=1
    idx=np.argsort(np.random.random(newx.shape[0]))
    newx=newx[idx]
    newy=newy[idx]
    np.save("/home/kasia/Deep_learning/Datasets/cifar10_aug_train_images.npy", newx)
    np.save("/home/kasia/Deep_learning/Datasets/cifar10_aug_train_labels.npy",newy)

    x=np.load("/home/kasia/Deep_learning/Datasets/cifar10_test_images.npy")
    newx=np.zeros((x.shape[0], dim, dim, 3), dtype="uint8")
    for i in range(x.shape[0]):
        im=Image.fromarray(x[i,:])
        im=im.crop((z,z,32-z,32-z))
        newx[i,...]=np.array(im)
    np.save("/home/kasia/Deep_learning/Datasets/cifar10_aug_test_images.npy", newx)

main()
```
