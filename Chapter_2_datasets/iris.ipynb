```python
import numpy as np
```

```python
# this loads the data and remove /n character
with open("/home/kasia/Deep_learning/Datasets/iris.data") as f:
    lines=[i[:-1] for i in f.readlines()]
```

```python
n=["Iris-setosa", "Iris-versicolor", "Iris-virginica"]
x=[n.index(i.split(",")[-1]) for i in lines if i!=""]
x=np.array(x, dtype="uint8")
```

```python
y=[[float(j) for j in i.split(",")[:-1]] for i in lines if i!=""]
y=np.array(y)
```

```python
i=np.argsort(np.random.random(x.shape[0]))
x=x[i]
y=y[i]
```

```python
np.save("/home/kasia/Deep_learning/Datasets/iris_features.npy",y)
np.save("/home/kasia/Deep_learning/Datasets/iris_labels.npy",x)
```
