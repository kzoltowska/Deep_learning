```python
import numpy as np
import keras
from keras.datasets import mnist
datasets="/home/kasia/Deep_learning/Datasets/"
```

```python
(xtrn, ytrn),(xtst, ytst)=mnist.load_data()
idx=np.argsort(np.random.random(ytst.shape[0]))
xtst=xtst[idx]
ytst=ytst[idx]
idx=np.argsort(np.random.random(ytrn.shape[0]))
xtrn=xtrn[idx]
ytrn=ytrn[idx]
np.save(datasets+"mnist_train_images.npy", xtrn)
np.save(datasets+"mnist_test_images.npy", xtst)
np.save(datasets+"mnist_train_labels.npy", ytrn)
np.save(datasets+"mnist_test_labels.npy", ytst)
```

```python
#convert to vectors
xtrnv=xtrn.reshape((60000,28*28))
xtstv=xtst.reshape((10000,28*28))
np.save(datasets+"mnist_train_vectors.npy", xtrnv)
np.save(datasets+"mnist_test_vectors.npy", xtstv)
```

```python
# generate permuted dataset
idx=np.argsort(np.random.random(28*28))
for i in range(60000):
    xtrnv[i,:]=xtrnv[i,idx]
for i in range(10000):
    xtstv[i,:]=xtstv[i,idx]
np.save(datasets+"mnist_train_scrambled_vectors.npy", xtrnv)
np.save(datasets+"mnist_test_scrambled_vectors.npy", xtstv)
```

```python
# Create scrambled images from the scambled vectors
t=np.zeros((60000,28,28))
for i in range(60000):
    t[i,:,:]=xtrnv[i,:].reshape((28,28))
np.save(datasets+"mnist_train_scambled_images.npy",t)

t=np.zeros((10000,28,28))
for i in range(10000):
    t[i,:,:]=xtstv[i,:].reshape((28,28))
np.save(datasets+"mnist_test_scambled_images.npy",t)
```
