### Buiding iris dataset for deeplearning

```python
import numpy as np
import matplotlib.pyplot as plt
d = np.load("/home/kasia/Deep_learning/Datasets/iris_train_features_augmented.npy")
l = np.load("/home/kasia/Deep_learning/Datasets/iris_train_labels_augmented.npy")

# there are 3 classes in iris, here we select only 1 and 2 for binary classification
d1 = d[np.where(l==1)]
d2 = d[np.where(l==2)]
```

```python
a=len(d1)
b=len(d2)
# create numpy array of zeros to later fill it in
x = np.zeros((a+b,2))

# we select features 2 and 3 and put them in x
x[:a,:] = d1[:,2:]
x[a:,:] = d2[:,2:]
```

```python
# create labels - recode them to 0 and 1
y = np.array([0]*a+[1]*b)

# get indeces to scramble the order
i = np.argsort(np.random.random(a+b))
x = x[i]
y = y[i]

np.save("/home/kasia/Deep_learning/Datasets/iris2_train.npy", x)
np.save("/home/kasia/Deep_learning/Datasets/iris2_train_labels.npy", y)
```

```python
# plot the data
i0 = np.where(y==0)[0]
i1 = np.where(y==1)[0]
x0 = x[i0]
x1 = x[i1]
plt.plot(x0[:,0],x0[:,1],marker='o',color='k',fillstyle='none',linestyle='none')
plt.plot(x1[:,0],x1[:,1],marker='^',color='k',fillstyle='none',linestyle='none')
plt.xlabel("$x_0$")
plt.xlabel("$x_1$")
plt.tight_layout(pad=0, w_pad=0, h_pad=0)
```

```python
# same code, but now it is repeated on the test set

d = np.load("/home/kasia/Deep_learning/Datasets/iris_test_features_augmented.npy")
l = np.load("/home/kasia/Deep_learning/Datasets/iris_test_labels_augmented.npy")
d1 = d[np.where(l==1)]
d2 = d[np.where(l==2)]
a=len(d1)
b=len(d2)
x = np.zeros((a+b,2))
x[:a,:] = d1[:,2:]
x[a:,:] = d2[:,2:]
y = np.array([0]*a+[1]*b)
i = np.argsort(np.random.random(a+b))
x = x[i]
y = y[i]
np.save("/home/kasia/Deep_learning/Datasets/iris2_test.npy", x)
np.save("/home/kasia/Deep_learning/Datasets/iris2_test_labels.npy", y)
```

```python
# plot the data
i0 = np.where(y==0)[0]
i1 = np.where(y==1)[0]
x0 = x[i0]
x1 = x[i1]
plt.plot(x0[:,0],x0[:,1],marker='o',color='k',fillstyle='none',linestyle='none')
plt.plot(x1[:,0],x1[:,1],marker='^',color='k',fillstyle='none',linestyle='none')
plt.xlabel("$x_0$")
plt.xlabel("$x_1$")
plt.tight_layout(pad=0, w_pad=0, h_pad=0)
```
