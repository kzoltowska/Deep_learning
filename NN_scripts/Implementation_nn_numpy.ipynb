```python
import numpy as np
import pickle
import sys 
```

```python

# define sigmoid function - it was used as an activation function in the mlpc
def sigmoid(x):
    return 1.0 / (1.0 + np.exp(-x))

# running the model using weights defined earlier during training - manual numpy implementation
def evaluate(x, y, w): 
    w12,b1,w23,b2,w34,b3 = w 
    nc = nw = 0 
    prob = np.zeros(len(y))
    for i in range(len(y)):
        a1 = sigmoid(np.dot(x[i], w12) + b1) 
        a2 = sigmoid(np.dot(a1, w23) + b2) 
        prob[i] = sigmoid(np.dot(a2, w34) + b3)[0] 
        z  = 0 if prob[i] < 0.5 else 1
        if (z == y[i]):
            nc += 1
        else:
            nw += 1
    return [float(nc) / float(nc + nw), prob]

xtest = np.load("/home/kasia/Deep_learning/Datasets/iris2_test.npy")
ytest = np.load("/home/kasia/Deep_learning/Datasets/iris2_test_labels.npy")
weights = pickle.load(open("/home/kasia/Deep_learning/Datasets/iris2_weights.pkl","rb"))
score, prob = evaluate(xtest, ytest, weights)
print()
for i in range(len(prob)):
    print("%3d:  actual: %d  predict: %d  prob: %0.7f" % (i, ytest[i], 0 if (prob[i] < 0.5) else 1, prob[i]))
print("Score = %0.4f" % score)
```
